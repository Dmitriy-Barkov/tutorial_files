### Определение периода с наибольшим количеством работ

```python
dep6_cat_response = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=6&q=cat')
cats = dep6_cat_response.json()
cats_IDs = cats['objectIDs'][:1000]
len(cats_IDs)
```

```python
required_columns = [
    'objectID', 
    'title', 
    'artistDisplayName', 
    'department', 
    'objectBeginDate', 
    'objectEndDate',
    'period',
    'objectName',
    'culture',
]

cat_paintings_df = pd.DataFrame()
for object_id in cats_IDs:
    response = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{object_id}')
    response_json = response.json()
    values_list = []
    values_dict = {}
    for data_key in required_columns:
        try:
            values_dict[data_key] = [response_json[data_key]]
        except:
            print(f'object_id - {object_id} could not be found in database')
            break
    current_df = pd.DataFrame(data=values_dict)
    cat_paintings_df = pd.concat([cat_paintings_df, current_df], ignore_index=True)
```
