#### Функция поиска родительской директории

```python
def find_parent_directory(path, level=1):
    upper_directory_path = path
    for i in range(level):
        upper_directory_path = os.path.dirname(upper_directory_path)
    return upper_directory_path
```

```python
df.groupby(['borough', 'hday'], as_index=False) \
    .agg({'pickups': 'mean'}) \
    .sort_values('pickups', ascending=False) \
```

```python
pickups_by_mon_bor = df.groupby(['pickup_month', 'borough'], as_index=False) \
                        .agg({'pickups': 'sum'}) \
                        .sort_values('pickups', ascending=False)
pickups_by_mon_bor
```

```python
def temp_to_celcius(fahrenheit_values):
    celsius = ((fahrenheit_values - 32) * 5) / 9
    return celsius

```
