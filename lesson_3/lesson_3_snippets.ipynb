```python
import pandas as pd
```

```python
df_input = pd.read_csv('/home/jupyter-d-barkov/Homework/lesson_3/lesson_3_data_1_.csv', encoding='windows-1251')
```

```python
user_df = df_input[['tc', 'art_sp']]
user_df.head()
```

```python
user_df = user_df.rename(columns={'tc': 'user_id', 'art_sp': 'brand_info'})
```

```python
user_df['brand_name'] = user_df.brand_info.apply(lambda x: x.split(' ')[-1])
```

```python
user_df
```

```python
user_purchases = user_df.groupby('user_id', as_index=False) \
                        .agg({'brand_name': 'count'}) \
                        .rename(columns={'brand_name': 'purchases'})
user_purchases
```

#### Выведем пользователей, которые совершили больше 5 покупок (входят в 75-%)

```python
user_purchases = user_df.groupby('user_id', as_index=False) \
                        .agg({'brand_name': 'count'}) \
                        .rename(columns={'brand_name': 'purchases'}) \
                        .query('purchases >= 5')
user_purchases
```
